```python
import pandas as pd
import numpy as np
df=pd.read_csv('inaturalistpoints_2088902879760696442.csv')
df_cleaned=df.dropna()
df_aedes=df_cleaned.loc[df_cleaned['ObsTaxonName']=='Aedes aegypti']
import re
month=[]
day=[]
year=[]
for i in range(len(df_aedes)):
    data_string=df_aedes['observationResCatObsPheTime'].values[i]
    matches = re.findall(r'\d+', data_string)  # Find all numbers
    month.append(int(matches[0]))
    day.append(int(matches[1]))
    year.append(int(matches[2]))
day=np.array(day)
month=np.array(month)
year=np.array(year)
df_aedes.insert(0,'day',day)
df_aedes.insert(1,'month',month)
df_aedes.insert(2,'year',year)
columns_list=['day','month','year','ObsTaxonName','latitude','longitude']
df_aedes_inaturalist=df_aedes.loc[:,columns_list]
df_aedes_inaturalist
```

```python
import pandas as pd
import numpy as np
df_mosquito_alert=pd.read_csv('mosquitoAlert_2750354792241070964.csv')
df_aedes=df_mosquito_alert.loc[df_mosquito_alert['Indentified by Human']=='Aedes aegypti']
import re
month=[]
day=[]
year=[]
for i in range(len(df_aedes)):
    data_string=df_aedes['observationResCatObsPheTime'].values[i]
    matches = re.findall(r'\d+', data_string)  # Find all numbers
    month.append(int(matches[0]))
    day.append(int(matches[1]))
    year.append(int(matches[2]))
day=np.array(day)
month=np.array(month)
year=np.array(year)
df_aedes.insert(0,'day',day)
df_aedes.insert(1,'month',month)
df_aedes.insert(2,'year',year)
columns_list=['day','month','year','Indentified by Human','latitude','longitude']
df_aedes_mosquito_alert=df_aedes.loc[:,columns_list]
df_aedes_mosquito_alert



    
```

```python
df_joint=pd.concat([df_aedes_inaturalist,df_aedes_mosquito_alert],axis=0,ignore_index=True)
df_joint['Specie']=df_joint['ObsTaxonName'].str.cat(df_joint['Indentified by Human'],sep=' ',na_rep=' ')
df_joint_specie=df_joint.drop(columns=['ObsTaxonName','Indentified by Human'])
df_joint_specie
```

```python
df_joint_specie.to_csv('GMOD_database_aegypti.csv')
```
