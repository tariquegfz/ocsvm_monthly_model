```python
import pandas as pd
import xarray as xr
import numpy as np
mosquito_climate_data=pd.read_csv('GMOD_aegypti_climate_data.csv')
mosquito_climate_data_original=pd.read_csv('GMOD_aegypti_climate_data.csv')
pick_year=2015
mosquito_climate_data_pyear=mosquito_climate_data_original.loc[mosquito_climate_data_original['year']==pick_year]
nm_data=pd.read_csv('LU_POP_2015-2019_31km_WGS84.csv') # Longitude is from -179 to 179
nm_pyear=nm_data.loc[nm_data['year']==pick_year]
climate_data_copy=mosquito_climate_data_pyear.copy()
```

```python
#climate_data_copy=climate_data.copy()
import geopandas as gpd
from shapely import wkt
for i in range(len(nm_pyear)):
    shapes=nm_pyear.iloc[i,15]
    polygon = wkt.loads(shapes)
    coords = np.dstack(polygon.boundary.xy).tolist()[0][:-1]
    longitude_range=[coords[x][0] for x in range(len(coords))]
    latitude_range=[coords[x][1] for x in range(len(coords))]
    latitude_max=np.max(latitude_range)
    latitude_min=np.min(latitude_range)
    longitude_max=np.max(longitude_range)
    longitude_min=np.min(longitude_range)
    condition=(climate_data_copy['latitude']>latitude_min) & (climate_data_copy['latitude']<latitude_max) & (climate_data_copy['longitude']>longitude_min) & (climate_data_copy['longitude']<longitude_max)
    matching_row=climate_data_copy.loc[condition]
    if matching_row.empty==False:
    
        climate_data_copy.loc[matching_row.iloc[0,0],'Population_Density']=nm_pyear.iloc[i,14]
        climate_data_copy.loc[matching_row.iloc[0,0],'land_use_0']=nm_pyear.iloc[i,4]
        climate_data_copy.loc[matching_row.iloc[0,0],'land_use_11']=nm_pyear.iloc[i,5]
        climate_data_copy.loc[matching_row.iloc[0,0],'land_use_22']=nm_pyear.iloc[i,6]
        climate_data_copy.loc[matching_row.iloc[0,0],'land_use_33']=nm_pyear.iloc[i,7]
        climate_data_copy.loc[matching_row.iloc[0,0],'land_use_44']=nm_pyear.iloc[i,8]
        climate_data_copy.loc[matching_row.iloc[0,0],'land_use_55']=nm_pyear.iloc[i,9]
        climate_data_copy.loc[matching_row.iloc[0,0],'land_use_66']=nm_pyear.iloc[i,10]
        climate_data_copy.loc[matching_row.iloc[0,0],'land_use_77']=nm_pyear.iloc[i,11]
        climate_data_copy.loc[matching_row.iloc[0,0],'geometry']=nm_pyear.iloc[i,15]
climate_data_subset=climate_data_copy.dropna()
climate_data_subset
```

```python
climate_data_subset=climate_data_subset.drop(columns='Unnamed: 0')
climate_data_subset
filename=str(pick_year)+"_GMOD_aegypti_climate_data_subset_pop_den_land_use_lat_lon.csv"
climate_data_subset.to_csv(filename)
```
